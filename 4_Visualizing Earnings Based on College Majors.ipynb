# Visualizing Earnings Based On College Majors

Exploring the job outcomes of students who graduated from college between 2010 and 2012. Will be applying the matplotlib to help visualize the data as we explore various aspects of the data set.



```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')

recent_grads.iloc[0]
```

```python
print(recent_grads.head())
print(recent_grads.tail())
```

```python
recent_grads.describe()
```

```python
raw_data_count = recent_grads.shape
print(raw_data_count)
```

```python
recent_grads = recent_grads.dropna()
```

```python
cleaned_data_count = recent_grads.shape
print(raw_data_count)
print(cleaned_data_count)
```

## Pandas, Scatter Plots

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', 
                  title='Median vs Sample size')
```

Looking at the median salary vs sample size, there is large amounts of data points at the intersection 500 samples and 40,000 median salaries. There is one outlier that is above 100,000 and another outlier at 40,000 with respect to ~4,500 samples.

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', 
                  title='Unemployment Rate vs Sample Size')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', 
                  title='Median vs Full Time')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', 
                  title='Unemployment Rate vs Share of Women ')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', 
                  title='Median vs Men')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', 
                  title='Median vs Women')
```
